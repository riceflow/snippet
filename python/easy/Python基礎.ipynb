# データフレーム関連

## 基礎

### データフレーム新規作成

- df = pd.concat([df, s1, s2], axis=1) # dfにs1とs2を列追加
- 参考リンク：[pandas.DataFrameに列や行を追加](https://note.nkmk.me/python-pandas-assign-append/)

### データ抽出

- df['A'] # 列抽出
- df.A # 列抽出

### データを一致したものを抽出[数値]

- df[df['age'] < 25]
- df.query('age < 25') # 上の別記法
- See. [pandas.DataFrameの行を条件で抽出するquery](https://note.nkmk.me/python-pandas-query/)

### データを一致したものを抽出[文字列]

- df1[df1['フィールド名'].str.contains('完全一致')] # str.containsでブール値取得し、それを再代入
- df1[df1['フィールド名'].str.contains('.*部分一致.*')] # 正規表現で書けばOK
- See. [pandas.DataFrameの行を条件で抽出するquery](https://note.nkmk.me/python-pandas-query/)

### インデックスの再作成

- df1.reset_index(drop=True)
- drop=Trueで削除した列には、再度インデックス作成する時には除外(デフォルト：False)
- See. [PandasでIndexをリセットするreset_index関数の使い方](https://deepage.net/features/pandas-reset-index.html)

### 欠損値の処理

- df1.dropna(subset=['hoge1, hoge2'])) # 列hoge1とhoge2のいずれかがNanであれば削除
- df1.dropna(how='all') # 全ての行が欠損していれば削除(デフォルトany(いずれか))
- df1.fillna(method='ffill') # 欠損値は置換(ffillで前の値と置換、bfillで後の値と置換)
- df1.fillna(df1の1行分) # NaNを含む対応関係にある列で固定値補完
- df1.fillna(df.mean()) # 上記の応用、各種計算値で埋める
- See.
  - [pandasで欠損値NaNを除外（削除）・置換（穴埋め）・抽出](https://note.nkmk.me/python-pandas-nan-dropna-fillna/)  
  - [Pandasでnan値を削除、穴埋めするfillna、dropnaの使い方](https://deepage.net/features/pandas-manipulate-na.html)
