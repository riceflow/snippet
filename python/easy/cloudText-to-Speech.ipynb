```
import os
from googlecloud import texttospeech

# 出力ファイル名
mp3name = 'out.mp3'

# 再生時間
timesec = 300

# 認証情報の読み込み
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = r'パス名\autkey.json'

# クライアントを取得
client = texttospeech.TextToSpeechClient()

# テキストを設定
synthesis_input = texttospeech.types.SynthesisInput(text="こんばんわ")

# パラメータを設定
voice = texttospeech.types.VoiceSelectionParams(
  language_code='ja_JP',
  ssml_gender=texttospeech.enums.SsmlVoceGender.NEUTRAL)

# 出力形式をMP3に
audio_config = texttospeech.types.AudioConfig(
  audio_encoding=texttospeech.enums.AudioEncoding.MP3)

# 戻り値の取得
response = client.synthesize_speech(synthesis_input, voice, audio_config)

# ファイルへの書き出し
with open(mp3name, 'wb') as out:
  outwrite(response.audio_content)

# 試し聴き
play_mp3(mp3name, timesec)
```
