# キーボードショートカット

## セル関連

- 上部にコードセルを追加：Ctrl + M → A
- 下部にコードセルを追加：Ctrl + M → B
- 選択しているセルの削除：Ctrl + M → D
- テキストセルに変更： Ctrl + M → M
- コードセルに変更： Ctrl + M → Y
- 全てのセルを開く：Ctrl + [
- 全てのセルを閉じる：Ctrl + ]

## セル移動
- カーソルを次のセルに移動：Ctrl + M → N
- カーソルを前のセルに移動：Ctrl + M → P
- 選択したセルを下に移動：Ctrl + M → J
- 選択したセルを上に移動：Ctrl + M → K
- セルの選択解除：Esc

## 実行関連
- 現在のセルを実行：Ctrl + Enter
- 現在のセルより上を実行：Ctrl + F8
- 現在のセルとそれ以降を実行：Ctrl + F10
- すべてのセルを実行：Ctrl + F9
- 実行の中断：Ctrl + M → I (interruptのi)
- 実行結果の表示/非表示：Ctrl + M → O


## その他

- チートシート：Ctrl + M → H
- コマンドパレット：Ctrl + Shift + P
- 行番号表示/非表示：Ctrl + M → L

```
# html の タブの情報を取得
from selenium.webdriver.common.by import By

# 次にクリックしたページがどんな状態になっているかチェックする
from selenium.webdriver.support import expected_conditions as EC

# 待機時間を設定
from selenium.webdriver.support.ui import WebDriverWait

# サンプルコード
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',options=options)
driver.implicitly_wait(10)
driver.get("https://www.yahoo.co.jp/")

# 検索テキストボックスに文字入力
input_seek = driver.find_element_by_id("srchtxt")
input_seek.clear()
input_seek.send_keys("テスト")
print("URL1=", driver.current_url)

# 検索ボタンを押下
driver.find_element_by_id("srchbtn").click()

# 要素の検証が終わるまで 10 秒待つ
element = WebDriverWait(driver, 10).until(
  EC.presence_of_element_located((By.ID, 'WS2br'))
)

# 結果を出力
print("URL2=", driver.current_url)
html = driver.page_source.encode('utf-8') #画面遷移したら、既にHTMLはdriverが保持
soup = BeautifulSoup(html, "html.parser")
print(soup.prettify())
```
